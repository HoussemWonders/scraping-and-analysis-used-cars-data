```python
from bs4 import BeautifulSoup
import requests
import pprint
import pandas as pd
```

```python
page = 1
l = []

while page <= 10:
    url = f"https://www.lacentrale.fr/listing?makesModelsCommercialNames=&options=&page={page}"
    doc = BeautifulSoup(requests.get(url).text, "html.parser")
    names = doc.find_all(class_="Text_Text_text Vehiculecard_Vehiculecard_title Text_Text_subtitle2")
    characteristics = doc.find_all(class_="Text_Text_text Vehiculecard_Vehiculecard_characteristicsItems Text_Text_body2")
    sublists = [characteristics[i:i + 4] for i in range(0, len(characteristics), 4)]
    prices = doc.find_all(text=" €")
    car_price = []

    for price in prices:
        parent = price.parent
        car_price.append(parent.get_text(strip=True))

    for i in range(15):
        d = {}
        d['name'] = names[i].string
        d['year'] = sublists[i][0].string
        d['kilometrage'] = sublists[i][1].string
        d['boite_de_vitesse'] = sublists[i][2].string
        d['moteur'] = sublists[i][3].string
        d['price'] = car_price[i]
        l.append(d)

    page += 1

pprint.pprint(l)
```

```python
df= pd.DataFrame(l)
```
